```python
from utils.data import get_data

import torch
import torch.nn as nn
import torch.optim as optim

from gru.model import RateGRU
from gru.trainer import Trainer
```

```python
# train a gru with the best params for the early learner
early_hparams = {'learning_rate': 0.0045105, 'batch_size': 16, 'hidden_dim': 128}

learner = "early"
data_early = get_data(learner)

config = {
        'lr': early_hparams['learning_rate'],
        'optimizer': optim.AdamW,
        'criterion': nn.MSELoss
}
    
print_every = 1
max_epochs = 1000
ckp_name = f'{learner}_best'

# instantiate model and trainer
gru_early = RateGRU(input_size=data_early['X_train'].shape[-1], 
                hidden_size=early_hparams['hidden_dim'], 
                output_size=data_early['Y_train'].shape[-1], 
                num_layers=1
            )

trainer_early = Trainer(gru_early, data_early,
                optim_config=config, 
                batch_size=early_hparams['batch_size'],
                n_epochs=max_epochs, 
                checkpoint_name=ckp_name, 
                print_every=print_every,
                verbose=True
                )
# train and eval
trainer_early.train()  
r2_val = trainer_early.best_val_r2
```
